For each user $j$, learn a parameter $\theta^{(j)}$. Predict user $j$ as rating movie $i$ with $(\theta^{(j)})^Tx^{(i)} $ 

$r(i,j) = 1$ if user j has rated movie $i$, otherwise 0

$y^{(i,j)} = $ rating by user $j$ of movie $i$ if $r(i,j) = 1$

$\theta^{(j)} = $ parameter vector for user $j$

$x^{(i)} = $ feature vector of movie $i$

$m^{(j)} = $ number of movies rated by user $j$

To learn $\theta$ :

$$\min\limits_{\theta^{(1)},...,\theta^{(n_u)}} ~~~\frac 1 {2} \sum\limits_{j = 1}^{n_u} \sum\limits_{i:r(i,j) = 1} \left((\theta^{(j)})^Tx^{(i)} - y^{(i,j)}\right)^2 +\frac {\lambda} {2} \sum\limits_{j = 1}^{n_u} \sum\limits_{k=1}^n(\theta_k^{(j)})^2$$




#### Gradient descent Update:

$\theta_k^{(j)} := \theta_k^{(j)} - \alpha \sum\limits_{i:r(i,j) = 1} ((\theta^{(j)})^Tx^{(i)}- y^{(i,j)})x_k^{(i)}~~~$       (for $k=0)$

$\theta_k^{(j)} := \theta_k^{(j)} - \alpha \left(\sum\limits_{i:r(i,j) = 1} ((\theta^{(j)})^Tx^{(i)}- y^{(i,j)})x_k^{(i)} +\lambda\theta_k^{(j)} \right)~~~$       (for $k>0)$

#### Optimizaton Algorithm

Given $\theta^{(1)},...,\theta^{(n_u)}$, to learn $x^{(i)}$:

$$\min\limits_{x^{(i)},...,x^{(n_m)}} ~~~\frac 1 {2} \sum\limits_{j = 1}^{n_m} \sum\limits_{j:r(i,j) = 1} \left((\theta^{(j)})^Tx^{(i)} - y^{(i,j)}\right)^2 +\frac {\lambda} {2} \sum\limits_{j = 1}^{n_m} \sum\limits_{k=1}^n(x_k^{(j)})^2$$

Minimizing $\theta^{(1)},...,\theta^{(n_u)}$ and $x^{(1)},...,x^{(n_m)}$ simultaneously:

$$\min\limits_{x^{(i)},...,x^{(n_m)},\theta^{(1)},...,\theta^{(n_u)}} ~~~\frac 1 {2}  \sum\limits_{(i,j):r(i,j) = 1} \left((\theta^{(j)})^Tx^{(i)} - y^{(i,j)}\right)^2$$

#### Collaborative Filtering Algorithm

1.) Initialize $x^{(i)},...,x^{(n_m)},\theta^{(1)},...,\theta^{(n_u)}$ to small random values

2.) Do Gradient descent:

$\theta_k^{(j)} := \theta_k^{(j)} - \alpha \left(\sum\limits_{i:r(i,j) = 1} ((\theta^{(j)})^Tx^{(i)}- y^{(i,j)})x_k^{(i)} +\lambda\theta_k^{(j)} \right)~~~$ 

$x_k^{(i)} := x_k^{(i)} - \alpha \left(\sum\limits_{j:r(i,j) = 1} ((\theta^{(j)})^Tx^{(i)}- y^{(i,j)})\theta_k^{(j)} +\lambda x_k^{(i)} \right)~~~$ 

3.) We no longer need $x_0 =1$, because the algorithm can make one if needed. Therefore, $x \in \mathbb {R}^n$ and $\theta \in \mathbb R^n$

```python
def do_grad(data,features, epochs = 1,alpha = 0.001, lmda = 1):
    x = randn(features,len(data.columns))
    theta = randn(len(data.index),features)

    for i in range(epochs):
        zee = (dot(theta,x) - data).fillna(0).values
        theta -= alpha*(dot(zee,x.T) + lmda*theta)
        x -= alpha*(dot(zee.T,theta).T + lmda*x)
        
    return (x,theta)    
seed(123)
scrubbed = temp - temp.mean()
do_grad(scrubbed,10,4)    
```
