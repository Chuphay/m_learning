# Unsupervised Learning

#### Clustering Algorithm

<b>K-means algorithm</b>

Input:
    
 - K (number of clusters)
    
 - Training set {$x^{(1)},x^{(2)}, ..., x^{(m)}$}
 
$x^{(i)} \in \mathbb{R}^n$ (drop $x_0=1$ by convention)

Randomly initialize K cluster centroids $\mu_1,\mu_2,...,\mu_K \in \mathbb{R}^n$$


Repeat {

$~~$for $i=1$ to $m$

$~~~~~~c^{(i)} :=$ index (from 1 to K) of cluster centroid closest to $x^{(i)}$

$~~~~~~\min\limits_k ~~||x^{(i)}-\mu_k||^2$

$~~$for $k=1$ to K

$~~~~~~\mu_k :=$ average (mean) of points assigned to cluster $k$

}

#### K-means optimization objective

$$J(c^{(1)},...,c^{(m)},\mu_1,...,\mu_K) = \frac 1 m \sum_{i=1}^m ||x^{(i)}-\mu_{c^{(i)}}||^2$$

where $\mu_{c^{(i)}} =  $ cluster centroid of cluster to which example $x^{(i)}$ has been assigned

For example, if $x^{(i)}$ has been assigned to 5, then $c^{(i)}=5$ and then $\mu_{c^{(i)}} = \mu_5$

#### Random initialization

Should have K $\leq m$

Randomly pick K training examples as your K centroids

```python
def k_means(data, K, depth = 6):
    """takes data like array([[3.48234593, 1.43069667],[1.13425727, 2.75657385],[3.59734485, 2.1155323]])
    which in this example is 3 data points with two features
    K is the number of clusters, depth is the depth that the algorithm will recur
    The program will return the centroids and the associated cost
    """
    example = arange(len(data))
    shuffle(example)
    example = example[:K]
    centroids = []
    for i in range(K):
        centroids.append(data[i].tolist())
    centroids = array(centroids)
    c = [0 for i in range(len(data))]
    for times in range(depth):
        for i,values in enumerate(data):
            temp = infty
            for k in range(len(centroids)):
                distance_sq = dot(centroids[k]-values,centroids[k]-values)
                if distance_sq < temp:
                    temp = distance_sq
                    c[i] = k
        for k in range(len(centroids)):
            mean = [0,0]
            count = 0
            for i,values in enumerate(data):
                if c[i] == k:
                    mean += values
                    count += 1
            if count != 0:
                centroids[k] = array(mean)/count
            else:
                raise ValueError ('what happened?')
    cost = 0
    for i, values in enumerate(data):
        cost += dot(centroids[c[i]]-values,centroids[c[i]]-values)
        
    return centroids, cost/len(data)                 
```
