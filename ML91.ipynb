# Anomaly Detection

Dataset: {$x^{(1)},x^{(2)},...,x^{(m)}$}

Is $x_{test}$ anomalous?

Model $p(x)$

$p(x_{test}) \leq \epsilon \rightarrow$ flag anomaly

#### Gaussian Distribution

$p(x;\mu,\sigma^2) = \frac 1 {\sqrt{2\pi\sigma^2}} \exp\left(- \frac {(x-\mu)^2}{2\sigma^2}\right)$

```python
gaussian = lambda x,mu,sigma: (exp(-(x-mu)**2/(2*sigma**2)))/sqrt(2*pi*sigma**2)

figsize(15,7)
x = linspace(-10,10,200)
for i in range(6):
    subplot(2,3,i+1)
    mu, sigma = 0, (i+1)/3.
    plot(x, gaussian(x,mu,sigma))
         
    xlabel('sigma %s'%sigma)
    ylim(0,1)

    
```

Dataset: {$x^{(1)},x^{(2)},...,x^{(m)}$}

$\mu = \frac 1 m \sum_{i=1}^m x^{(i)}$

$\sigma^2 = \frac 1 m \sum_{i=1}^m  (x^{(i)} - \mu)^2$

Note that this definition of variance is different than the usual definition... but who really cares on big data sets

#### Density Estimation

$p(x) = \Pi_{j=1}^n p(x_j;\mu_j,\sigma_j^2)$

```python
seed(1234)
data = normal((2,2),(0.1,10),(100,2))
figsize(12,6)
subplot(121)
scatter(data[:,0],data[:,1])
epsilon = 0.1
for x in data:
    p = gaussian(x[0],2,0.1)*gaussian(x[1],2,10)
    if p<epsilon:
        scatter(x[0],x[1],c='r')
xlim(1.2,2.8)        
subplot(122)
epsilon = 0.01
for x in data:
    p = gaussian(x[0],2,0.1)*gaussian(x[1],2,10)
    if p<epsilon:
        scatter(x[0],x[1],c='r',marker = 'x') 
    else: 
        scatter(x[0],x[1],c = 'b')
xlim(1.2,2.8)         
show()        
```

#### Anomaly Detection with the Multivariate Gaussian

Fit model $p(x)$ by setting

$\mu = \frac 1 m \sum_{i=1}^m x^{(i)}$

$\Sigma = \frac 1 m \sum_{i=1}^m (x^{(i)}-\mu)(x^{(i)}-\mu)^T$

Given a new example $x$, compute

$p(x) = \frac 1 {(2\pi)^{\frac n 2}|\Sigma|^{\frac 1 2}} \exp \left( - \frac 1 2 (x-\mu)^T \Sigma^{-1} (x-\mu) \right)$

Flag an anamoly if $p(x) < \epsilon$
