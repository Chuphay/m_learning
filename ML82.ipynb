#### Principal Component Analysis (PCA)

Reduce from 2 dimensions to 1: find a direction (a vector ${u}^{(1)} \in \mathbb{R}^n$) onto which to project the data so as to minimize the projection error.

Reduce from n dimensions to k dimensions: find $k$ vectors ${u}^{(1)},{u}^{(2)},...,{u}^{(k)}$ onto which to project the data so as to minimize the projection error.

### PCA Algorithm

<b>Data Preprocessing</b>

Training set: $x^{(1)}, x^{(2)}, ..., x^{(m)} $

Preprocessing (feature scaling/ mean normalization)

$$\mu_j = \frac 1 m \sum_{i=i}^m x_j^{(i)}$$

Replace each $x_j^{(i)}$ with $x_j-\mu_j$

If different features on different scales, scale features to have comparable range of values

```python
scrubbed_boston = (boston - boston.mean())/(boston.max() - boston.min())
scrubbed_boston_test = (test_boston_data - test_boston_data.mean())/(test_boston_data.max() - test_boston_data.min())
scrubbed_data = (df - df.mean())/(df.max()-df.min())
```

```python
figsize(15,10)
scrubbed_boston.hist()
show()
```

<b>PCA Algorithm</b>

Reduce data from n dimensions to k dimensions

Compute covariance matrix:

$$\Sigma = \frac 1 m \sum_{i=1}^m (x^{(i)})(x^{(i)})^T$$


```python
def get_cov(df):
    sigma = 0
    for row in df.values:
        sigma += outer(row,row)
    return sigma/len(df.values)   

bos_cov = get_cov(scrubbed_boston)
data_cov = get_cov(scrubbed_data)
print data_cov == dot(scrubbed_data.values.T,scrubbed_data.values)/len(scrubbed_data.values)
```

Compute eigenvectors of $\Sigma$

```python
data_U, data_S, data_V = svd(data_cov)
bos_U, bos_S, bos_V = svd(bos_cov)
```

From [U,S,V] = svd(sigma), we get:

$U = \left[ u^{(1)} u^{(2)} ... u^{(n)} \right] ~~~ U \in \mathbb{R}^{n\times n}$ 

take the first $k$ columns of $U$

$$U_{reduce} = \left[ u^{(1)} u^{(2)} ... u^{(K)} \right] ~~~ U_{reduce} \in \mathbb{R}^{n\times k}$$

$$z = U_{reduce}^T \cdot x$$


```python
U_data_R = data_U[:,:2]
z_data =  dot(U_data_R.T,scrubbed_data.T)
U_bos_R = bos_U[:,:2]
z_boston = dot(U_bos_R.T,scrubbed_boston.T)
```

#### Choosing the number of principal components

Averaged square projection error: $\frac 1 m \sum_{i=1}^m ||x^{(i)} - x_{approx.}^{(i)}||^2$

Total Variation of the data: $\frac 1 m \sum_{i=1}^m ||x^{(i)}||^2$

Typically choose $k$ s.t.

$$\frac {\frac 1 m \sum _{i=1}^m ||x^{(i)} - x_{approx.}^{(i)}||^2}{\frac 1 m \sum _{i=1}^m ||x^{(i)}||^2} \leq 0.01$$

"99% of the variance is retained"

For a given $k$:

$$1-\frac {\sum_{i=1}^k S_i}{\sum_{i=1}^n S_i} \leq 0.01$$

so, for our boston data set, we have:

```python
for i in range(1,len(bos_S)+1):
    print sum(bos_S[:i])/sum(bos_S), 'k = ',i
```

So we'll use $k=8$ to get some compression, but not too much

```python
U_bos_R = bos_U[:,:8]
z_boston = dot(U_bos_R.T,scrubbed_boston.T)
```

#### Reconstruction from Compressed Representation 

$$x_{approx.} = U_{reduce} \cdot z $$

#### Applying PCA

supervised learning speedup:

$$((x^{(1)},y^{(1)}),(x^{(2)},y^{(2)}),...,(x^{(m)},y^{(m)}))$$

extract the inputs:

unlabeled dataset: $x^{(1)},x^{(2)},...x^{(m)} \in \mathbb{R}^n$

apply PCA: $\rightarrow z^{(1)},z^{(2)},...z^{(m)} \in \mathbb{R}^k$

New training set: $((z^{(1)},y^{(1)}),(z^{(2)},y^{(2)}),...,(z^{(m)},y^{(m)}))$



```python
from scipy.optimize import minimize
def train(data, target, lmda = 0.4):
    X= concatenate((ones((data.shape[0],1)),data), axis =1)
    X = asarray(X) 
    init = zeros(data.shape[1]+1)
    cost = lambda theta: ((dot(theta,X.T) -target)**2).sum() +lmda*((theta[1:])**2).sum()
    result = minimize(cost, init)
    print result['success']
    theta =  result['x']
    return theta

train(z_boston.T, boston_target.values)
```
