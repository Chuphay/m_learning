### Define a function for creating mapped features to non linear quadratics

```python
def map_features(X1, X2, depth):
    """
    Returns an array of features consisting of nonlinear
    relationships between two input arrays.
    """
    
    m = len(X1)
    
    # reshape input arrays to matrices:
    X1 = X1.reshape((m, 1))
    X2 = X2.reshape((m, 1))
    
    # Initialize the output array
    out = np.ones((m, 1))
    
    # remember that python indices start at 0
    for i in range(1, depth + 1):
        for j in range(i + 1):
            Xnew = (X1**(i - j) * X2**j)
            out = np.column_stack((out, Xnew))
    
    return out
```

### Here we define our sigmoid function which accepts integers and arrays:

```python
# integer operator
def f(z):
    if z < -100:
        return 0
    elif z > 100:
        return 1
    else:
        return 1 / (1 + np.exp(-z))

sigmoid = np.vectorize(f)
```

### Here we define our cost function and gradient function (Including regularization):

Recall that the corresponding differentials are:

\begin{eqnarray*}
\ \ \ \ \ \ \ \ \ \ 
\frac{\partial J(\theta)}{\partial \theta_0} & = & \frac{1}{m} \sum_{i = 1}^{m} \left( h_{\theta} (x^{(i)}) - y^{(i)} \right) x_0^{(i)} 
& \mbox{for}\ \ j = 0 \\
\ \ \ \ \ \ \ \ \ \ 
\frac{\partial J(\theta)}{\partial \theta_j} & = & \left( \frac{1}{m} \sum_{i = 1}^{m} \left( h_{\theta} (x^{(i)}) - y^{(i)} \right) x_j^{(i)} \right)
+ \frac{\lambda}{m}{\theta_j}\ \ \ \ 
& \mbox{for}\ \ j \geq 1
\end{eqnarray*}



Before we continue I define a function for calculating logrithms. This is to avoid a runtime error and handles being passed negative values. This way we can pass the entire cost function without returning an error.

```python
def f2(x):
    if x >= 0:
        return np.log(x)
    elif x == 0:
        return -9999
    else:
        return 0
    
ln = np.vectorize(f2)
```

```python
# regularized cost function
def compute_cost(theta, X, Y, lam=0):
    m, n = X.shape
    
    # reshape to account for variables passed as 1-dimensional arrays
    Y = Y.reshape((m, 1))
    theta = theta.reshape((n, 1))
    
    # calculate h_theta(x) in a vectorized format (z = X.dot(theta)):
    hx = sigmoid(X.dot(theta))
    
    # calculate J
    J = (1/m) * (np.sum((-Y * ln(hx)) - ((1 - Y) * ln(1 - hx))))
    #J += (lam/(2*m)) * theta[1:].T.dot(theta[1:]) # regularization term
    
    # calculate theta differentials
    grad = (1/m) * X.T.dot(hx - Y)
    #grad += np.vstack((np.array([0]), (lam/m) * theta[1:]))
    
    return J, grad
```

Use scipy.optimize to optimize our function:

```python
# first lets reload our dataset and initialize
data = np.loadtxt("ex2data1.txt", delimiter=',')
print data[:3]
Y = data[:, 2]
X1 = data[:, 0]
X2 = data[:, 1]
X = map_features(X1, X2, 1)
print X[:3]
m, n = X.shape
print m, n
lamb = 0.01
theta = np.zeros((n, 1))
print len(X), theta
```

```python
meth = 'L-BFGS-B', 'TNC', 'BFGS', 'CG', 'Newton-CG'
print("Optimize:")
print len(X), len(Y)
optimum = opt.minimize(
    compute_cost, theta,
    method=meth[3],
    jac=True, args=(X, Y)) #, options={'maxiter': 1500}

print("fun:     %s" % optimum.fun)
print("status:  %s" % optimum.status)
print("message: %s" % optimum.message)
print("success: %s" % optimum.success)
if optimum.success:
    print("Optimum theta(s):")
    print(optimum.x)
    theta_final = optimum.x
```

```python
# plot our descision boundary
u = np.linspace(-1, 1.5, 50).reshape((50, 1))
v = np.linspace(-1, 1.5, 50).reshape((50, 1))

print(u.shape)

z = np.zeros((len(u), len(v)))

# evaluate z = theta * x over the grid
for i in range(len(u)):
    for j in range(len(v)):
        z[i, j] = map_features(u[i], v[j], 6).dot(theta_final)

U, V = np.meshgrid(u, v)
X3 = np.column_stack((np.ones((m, 1)), X1.reshape((m, 1)), X2.reshape((m, 1))))
plot_data(X3, Y.reshape((m, 1)))
pl.contour(U, V, z, [0, 0])
```
