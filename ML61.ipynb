```python
seed(1234) 
z = lambda x: x**2 +5+2*randn()
x = linspace(0,5,10, endpoint = False)
data = []
for i in x:
    data.append(z(i))
figsize(7,5)
plot(x,data,'bo',label ='test set')
training,test = [],[]
for e,i in enumerate(x):
    zee = rand()
    if zee>0.7:
        test.append([i,data[e]])
    else:
        training.append([i,data[e]])
training = array(training)
test = array(test)

plot(training[:,0],training[:,1],'ro',label = 'training set')
legend(loc = 'best')
show()
```

#### Training/Testing Procedure for Linear Regression

Learn parameter $\theta$ from training data (minimize training error $J(\theta)$)

Compute test set error:

$$J_{test}(\theta) = \frac 1 {2m_{test}} \sum_{i=1}^{m_{test}} (h_{\theta}(x_{test}^{(i)})-y_{test}^{(i)})^2 $$

For Logistic regression, we use <b>Misclassification error</b> (0/1 misclassification error):

$$err(h_{\theta}(x),y) = 
\begin{cases} 
1 & if \hspace{2mm} h_{\theta}(x) \geq 0.5, ~ y=0 ~~ or~~~ h_{\theta}(x) \leq 0.5, ~ y=1 \\ 
0 & otherwise
\end{cases}$$

$$Test~error = \frac 1 {m_{test}} \sum_{i=1}^{m_{test}} err (h_{\theta}(x_{test}^{(i)}),y_{test}^{(i)})$$

```python
from scipy.optimize import minimize

def model(d,x,y, plot_on = False):
    X=[]
    for i in range(d+1):
        X.append(x**i)
    init = zeros(d+1)
    cost = lambda theta: ((dot(theta,X) -y)**2).sum()
    result = minimize(cost, init)
    theta =  result['x']
    if plot_on:
        z = linspace(x[0]-1, x[-1]+0.5,100)
        plot(x,y,'ro')
        X=[]
        for i in range(d+1):
            X.append(z**i)   
        plot(z,dot(theta,X))
        string = 'Degree = %d' %d  
        xlabel(string)
    return theta

figsize(15,10)
weights = {}
for i in range(7):
    if i == 0:
        weights[0] = model(0,training[:,0],training[:,1])
    else:
        num = '23' + str(i)
        subplot(int(num))
        weights[i] = model(i,training[:,0],training[:,1],plot_on = True)
        xlim(-1,5),ylim(0,30)

show()
```

And now we want to compute the error with our test set using:

$$J_{test}(\theta) = \frac 1 {2m_{test}} \sum_{i=1}^{m_{test}} (h_{\theta}(x_{test}^{(i)})-y_{test}^{(i)})^2 $$

```python
def error_test(d,theta,data):
    x,y = data[:,0], data[:,1]
    X=[]
    for i in range(d+1):
        X.append(x**i)  
    cost = ((dot(theta,X) -y)**2).sum()/len(x)
    return cost

for i in range(7):
    print 'error: %.5f  with degree %d' %(error_test(i,weights[i],test),i) 


```

#### Bias vs. Variance

```python
from scipy.interpolate import interp1d

x = arange(7)
y_train = []
y_test = []
for i in x:
    y_train.append(error_test(i,weights[i],training))
    y_test.append(error_test(i,weights[i],test))

figsize(7,5)
xnew = np.linspace(0, 5, 40)
f_train = interp1d(x, y_train, kind='cubic')
f_test = interp1d(x, y_test, kind='cubic')
plot(xnew,f_train(xnew))
plot(xnew,f_test(xnew))
ylim(0,20)
legend(['training error','test error'],loc= 'best')
show()
```

#### Regularization and Bias/Variance

Linear regression with regularization

$$h_{\theta}(x) = \theta_0 + \theta_1 x + \theta_2 x^2 + ...$$

$$J(\theta) = \frac 1 {2m} \left[ \sum_{i=1}^m (h_{\theta}(x^{(i)})-y^{(i)})^2 + \lambda \sum_{j=1}^m \theta_j^2 \right]$$ 

$$J_{train}(\theta) = \frac 1 {2m_{test}} \sum_{i=1}^{m_{train}} (h_{\theta}(x_{train}^{(i)})-y_{train}^{(i)})^2 $$

$$J_{test}(\theta) = \frac 1 {2m_{test}} \sum_{i=1}^{m_{test}} (h_{\theta}(x_{test}^{(i)})-y_{test}^{(i)})^2 $$

```python
def train(data, deg = 6, lmda = 0, plot_on = False):
    x,y = data[:,0], data[:,1]
    X=[]
    for i in range(deg+1):
        X.append(x**i)
    init = zeros(deg+1)
    cost = lambda theta: ((dot(theta,X) -y)**2).sum() +lmda*((theta[1:])**2).sum()
    result = minimize(cost, init)
    theta =  result['x']
    if plot_on:
        z = linspace(min(x)-1, max(x)+0.5,100)
        plot(x,y,'ro')
        X=[]
        for i in range(deg+1):
            X.append(z**i)   
        plot(z,dot(theta,X))
        string1 = 'Lamda = %f' %lmda 
        string2 = 'degree = %d' %deg
        ylabel(string1)
        xlabel(string2)
    return theta

figsize(18,12)
subplot(4,3,1)
weights[0] = train(training, lmda = 0, plot_on = True)
xlim(-1,5),ylim(0,30)
for i in range(100):
    lmda = 0.02*2**(i-1)
    if i<11:
        subplot(4,3,i+2)
        weights[i+1] = train(training, lmda = lmda, plot_on = True)
        
        xlim(-1,5),ylim(0,30)
    else:
        weights[i+1] = train(training, lmda = lmda)
```
